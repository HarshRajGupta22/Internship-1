after that you will reach to a webpage having a lot of sunglasses. From this page you can scrap the required data as usual.

after scraping data from the first page, go to the “Next” Button at the bottom of the page , then click on it.

repeat this until you get data for 100 sunglasses.

Note that all of the above steps have to be done by coding only and not manually.

```python
import selenium
import pandas as pd
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
# connecting Webdriver
driver = webdriver.Chrome(r'C:\chromedriver.exe')
driver
# Opening naukri.com
url ='https://www.flipkart.com/'
driver.get(url)
```

```python
driver.find_element_by_xpath('//div[@class="_3OO5Xc"]/input').send_keys('sunglasses')
driver.find_element_by_xpath('//button[@class="L0Z3Pu"]').click()
```

```python
#creating the empty lists
Brand=[]
Product_Description=[]
Price=[]
Discount=[]
```

```python
# Scraping URl of 100 Sunglasses
URL=[]
for i in range(0,3):
    S_g_url=driver.find_elements_by_xpath('//a[@class="_2UzuFa"]')
    for i in S_g_url:
        URL.append(i.get_attribute('href'))
    # Clicking on NEXT button at end of page
    driver.find_element_by_xpath('//a[@class="_1LKTO3"]').click()
    time.sleep(4)
```

```python
URL[:100]
```

```python
for i in URL[:100]:
    driver.get(i)
    try:
        brand=driver.find_element_by_xpath("//span[@class='G6XhRU']").text
        Brand.append(brand)
        description=driver.find_element_by_xpath("//span[@class='B_NuCI']").text
        Product_Description.append(description)
        price=driver.find_element_by_xpath("//div[@class='_30jeq3 _16Jk6d']").text
        Price.append(price)
        discount=driver.find_element_by_xpath("//div[@class='_3Ay6Sb _31Dcoz pZkvcx']/span").text
        Discount.append(discount)
    except NoSuchElementException:
        Brand.append("NaN")
        Product_Description.append("NaN")
        Price.append("NaN")
        Discount.append('NaN')
```

```python
Sunglasses=pd.DataFrame({})
Sunglasses['Brand']=Brand[:100]
Sunglasses['Product_Description']=Product_Description[:100]
Sunglasses['Price']=Price[:100]
Sunglasses['Discount']=Discount[:100]

Sunglasses
```

```python
Laptop_model = []
Price = []
URL = []
Rating = []
```

```python
# Scraping data for laptop titles
laptop_model = driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']")
for i in laptop_model:
    Laptop_model.append(i.text)
else:
    Laptop_model.append('-')

# Scraping laptop price
price = driver.find_elements_by_xpath("//span[@class='a-price-whole']")
for i in price:
    Price.append(i.text)
else:
    Price.append('-')
```

**Q10: Write a python program to scrape the salary data for Data Scientist designation.
You have to scrape Company name, Number of salaries, Average salary, Minsalary, Max Salary.
The above task will be, done as shown in the below steps:**

1. First get the webpage https://www.ambitionbox.com/
2. Click on the salaries option as shown in the image.
3. After reaching to the following webpage, In place of “Search Job Profile” enters “Data Scientist” and then click on “Data Scientist”. You have to scrape the data ticked in the above image.
4. Scrape the data for the first 10 companies. Scrape the company name, total salary record, average
salary, minimum salary, maximum salary, experience required.
5. Store the data in a dataframe.


```python
# connecting Webdriver
driver = webdriver.Chrome(r'C:\chromedriver.exe')
driver
#specifying the url of the webpage to be scraped
url="https://www.ambitionbox.com/"
driver.get(url)
time.sleep(3)
```

```python
# Extracting the element to click on the salaries option
Salary_button = driver.find_element_by_xpath('//*[@id="headerWrapper"]/nav/nav/a[4]')
Salary_button.click()
time.sleep(3)
```

```python
#Extracting the search element to enter data scientist designation
Search_bar = driver.find_element_by_xpath('//*[@id="jobProfileSearchbox"]')
Search_bar.send_keys("Data scientist")
time.sleep(1)

# clicking on data scientist designation using the extracted xpath
Datascientist = driver.find_element_by_xpath('//*[@id="salaries"]/main/section[1]/div[2]/div[1]/span/div/div/div[1]/div/div/p')
Datascientist.click()
time.sleep(2)
```

```python
comp_name = driver.find_elements_by_xpath('//div[@class="name"]/a')
total_salary = driver.find_elements_by_xpath('//div[@class="name"]/span')
avg_salary = driver.find_elements_by_xpath('//p[@class="averageCtc"]')
min_salary = driver.find_elements_by_xpath('//div[@class="value body-medium"]')
max_salary = driver.find_elements_by_xpath('//div[@class="value body-medium"]')
exp = driver.find_elements_by_xpath('//div[@class="salaries sbold-list-header"]')
```

```python
# Creating Empty list
Company_Title = []
Total_salary = []
Avg_salary = []
Min_salary = []
Max_salary = []
Experience = []
```

```python
for i in comp_name:
    Company_Title.append(i.text)
    
for i in total_salary:
    Total_salary.append(i.text)
    
for i in avg_salary:
    Avg_salary.append(i.text)
    
for i in range(0,len(min_salary),2):
    Min_salary.append(min_salary[i].text)
    
for i in range(1,len(min_salary),2):
    Min_salary.append(min_salary[i].text)

for i in exp:
    Experience.append(i.text.split()[3:])
```

```python
Salary_Data = pd.DataFrame({'Company Name':Company_Title[:10],'No. of Salaries':Total_salary[:10],'Average Salary':Avg_salary[:10],
                  'Minimum Salary':Min_salary[:10],'Maximum Salary':Min_salary[:10],'Experience Required':Experience[:10]})
Salary_Data
```
